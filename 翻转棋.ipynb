```python
import numpy as np
import pandas as pd
import matplotlib as plt
%matplotlib inline
```

```python
x_position = pd.Series(['000','001','010','011','100','101','110','111'])
y_position = pd.Series(['000','001','010','011','100','101','110','111'])
two_ten = {'000':0,'001':1,'010':2,'011':3,'100':4,'101':5,'110':6,'111':7}
```

```python
qipan = pd.DataFrame(np.zeros([8,8])).astype(int)
qipan.iloc[3,4]=1
qipan.iloc[4,3]=1
qipan.iloc[3,3]=2
qipan.iloc[4,4]=2
qipan
```

```python
qipan.rename(index=y_position,columns=x_position,inplace=True)
```

```python
def change(ele):
    if ele == 1:
        ele = 2
    else:
        ele = 1
    return ele

def checkout(x_pos,y_pos,color):
    x = two_ten[x_pos]
    y = two_ten[y_pos]
    inv = change(color)
    flag = 0
    x_l = x - 1
    x_r = x + 1
    y_u = y - 1
    y_d = y + 1
    x_ld = x_l
    y_ld = y_d
    x_ru = x_r
    y_ru = y_u
    x_lu = x_l
    y_lu = y_u
    x_rd = x_r
    y_rd = y_d
    while x_r<8:
        if qipan.iloc[y,x_r] == color:
            if (x+1) == x_r:
                break
            else:
                flag = 1
                qipan.iloc[y,x:(x_r+1)] = color
                break
        elif qipan.iloc[y,x_r] == inv:
            x_r += 1
        else:
            break
    # check the right side
    
    while x_l>-1:
        if qipan.iloc[y,x_l] == color:
            if (x-1) == x_l:
                break
            else:
                flag = 1
                qipan.iloc[y,(x_l-1):x] = color
                break
        elif qipan.iloc[y,x_l] == inv:
            x_l -= 1
        else:
            break
    # check the left side
    
    while y_d<8:
        if qipan.iloc[y_d,x] == color:
            if (y+1) == y_d:
                break
            else:
                flag = 1
                qipan.iloc[y:(y_d+1),x] = color
                break
        elif qipan.iloc[y_d,x] == inv:
            y_d += 1
        else:
            break
    # check the down side
         
    while y_u>-1:
        if qipan.iloc[y_u,x] == color:
            if (y-1) == y_u:
                break
            else:
                flag = 1
                qipan.iloc[(y_u-1):y,x] = color
                break
        elif qipan.iloc[y_u,x] == inv:
            y_u -= 1
        else:
            break
    # check the up side
    
    while x_ru<8 and y_ru>-1:
        if qipan.iloc[y_ru,x_ru] == color:
            if (x+1) == x_ru:
                break
            else:
                flag = 1
                cha = x_ru - x
                for ele in range(cha+1):
                    qipan.iloc[y-ele,x+ele] = color
                break
        elif qipan.iloc[y_ru,x_ru] == inv:
            x_ru += 1
            y_ru -= 1
        else:
            break
    # check the right-up side
    
    while x_ld>-1 and y_ld<8:
        if qipan.iloc[y_ld,x_ld] == color:
            if (x-1) == x_ld:
                break
            else:
                flag = 1
                cha =  x - x_ld
                for ele in range(cha+1):
                    qipan.iloc[y+ele,x-ele] = color
                break
        elif qipan.iloc[y_ld,x_ld] == inv:
            x_ld -= 1
            y_ld += 1
        else:
            break
    # check the left-down side

    while x_rd<8 and y_rd<8:
        if qipan.iloc[y_rd,x_rd] == color:
            if (x+1) == x_rd:
                break
            else:
                flag = 1
                cha = x_rd - x
                for ele in range(cha+1):
                    qipan.iloc[y+ele,x+ele] = color
                break
        elif qipan.iloc[x_rd,y_rd] == inv:
            x_rd += 1
            y_rd += 1
        else:
            break
    # check the right-down side
   
    while x_lu>-1 and y_lu>-1:
        if qipan.iloc[y_lu,x_lu] == color:
            if (x-1) == x_lu:
                break
            else:
                flag = 1
                cha =  x - x_lu
                for ele in range(cha+1):
                    qipan.iloc[y-ele,x-ele] = color
                break
        elif qipan.iloc[y_lu,x_lu] == inv:
            x_lu -= 1
            y_lu -= 1
        else:
            break
    # check the left-up side
    
    if flag==1:
        qipan.iloc[y,x] = color
        return 1
    else:
        return 0

now = 1 # Aiming to point whether it is to go white or black 
    
def readstr(strr):
    (y_pp,x_pp) = strr.replace('(','').replace(')','').split(',',maxsplit=1)
#     print(x_pp)
#     print(y_pp)
    return (x_pp,y_pp)
      
def run_go(strr):
    global now
    x,y = readstr(strr)
    x=str(x)
    y=str(y)
    tag = checkout(x,y,now)
    if tag == 1:
        if  now == 1:
            now =2
            print('Its turn for White to go')
        else:
            now = 1
            print('Its turn for Black to go')
    else:
        print('What you done is illeagle\n')
        if now==1:
            print('Its turn for Black to go')
        else:
            print('Its turn for White to go')
        
```
